```python
lockers = list(range(1, 101))
```

Creating two empty arrays open and close

```python
open=[]
close=[]
```

Student 2 starts close every second door, so starting with student 2 to close..

```python
for a in range(len(lockers)):
    if a%2==0:
        open.append(lockers[a])
    else:
        close.append(lockers[a])
print('opened lockers',open)
print('closed lockers',close)
```

and then every other student with respect to thier number does opposite of what is done on their position.., if close then open, if open then close.

```python
for i in range(3,len(lockers)+1):
    for j in range(i,len(lockers)+1):
            if j%i==0:
                if j in close:
                    close.remove(j)
                    open.append(j)
                elif j in open:
                    open.remove(j)
                    close.append(j)
print('open lockers',open)
```
